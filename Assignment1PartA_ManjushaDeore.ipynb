**Question A1.** Design and train a convolutional neural network from scratch for image
classification on the CIFAR-100 dataset. The network should be of moderate size and
computationally feasible to train using available resources. Clearly describe the architecture,
including the number of convolutional layers, kernel sizes, activation functions,
and pooling operations. Train the model and report the training and validation accuracy.
Comment on the convergence behavior and overall performance.





```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tqdm import tqdm

# ==========================================
# 1. Data Processing Pipeline
# ==========================================
def get_dataloaders(batch_size=128):
    """
    Downloads CIFAR-100, applies data augmentation to the training set,
    and returns DataLoader objects.
    """
    mean = (0.5071, 0.4865, 0.4409)
    std = (0.2673, 0.2564, 0.2762)

    transform_train = transforms.Compose([
        transforms.RandomCrop(32, padding=4),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(15),
        transforms.ToTensor(),
        transforms.Normalize(mean, std),
    ])

    transform_test = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize(mean, std),
    ])

    trainset = torchvision.datasets.CIFAR100(
        root='./data', train=True, download=True, transform=transform_train)
    trainloader = torch.utils.data.DataLoader(
        trainset, batch_size=batch_size, shuffle=True, num_workers=2)

    testset = torchvision.datasets.CIFAR100(
        root='./data', train=False, download=True, transform=transform_test)
    testloader = torch.utils.data.DataLoader(
        testset, batch_size=batch_size, shuffle=False, num_workers=2)

    return trainloader, testloader

# ==========================================
# 2. Model Architecture
# ==========================================
class CustomCNN(nn.Module):
    def __init__(self, num_classes=100):
        super(CustomCNN, self).__init__()

        self.conv1 = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )

        self.conv2 = nn.Sequential(
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )

        self.conv3 = nn.Sequential(
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )

        self.conv4 = nn.Sequential(
            nn.Conv2d(256, 512, kernel_size=3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )

        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(512 * 2 * 2, 1024),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(1024, num_classes)
        )

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = self.conv4(x)
        x = self.classifier(x)
        return x

# ==========================================
# 3. Training and Evaluation Functions
# ==========================================
def train_and_evaluate(model, trainloader, testloader, criterion, optimizer, scheduler, epochs, device):
    history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}

    for epoch in range(epochs):
        model.train()
        running_loss, correct, total = 0.0, 0, 0

        for inputs, labels in tqdm(trainloader, desc=f"Epoch {epoch+1}/{epochs} [Train]"):
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

        epoch_train_loss = running_loss / total
        epoch_train_acc = 100. * correct / total

        model.eval()
        running_val_loss, val_correct, val_total = 0.0, 0, 0

        with torch.no_grad():
            for inputs, labels in testloader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)

                running_val_loss += loss.item() * inputs.size(0)
                _, predicted = outputs.max(1)
                val_total += labels.size(0)
                val_correct += predicted.eq(labels).sum().item()

        epoch_val_loss = running_val_loss / val_total
        epoch_val_acc = 100. * val_correct / val_total

        history['train_loss'].append(epoch_train_loss)
        history['train_acc'].append(epoch_train_acc)
        history['val_loss'].append(epoch_val_loss)
        history['val_acc'].append(epoch_val_acc)

        scheduler.step()
        current_lr = scheduler.get_last_lr()[0]

        print(f"Train Loss: {epoch_train_loss:.4f} | Train Acc: {epoch_train_acc:.2f}% | "
              f"Val Loss: {epoch_val_loss:.4f} | Val Acc: {epoch_val_acc:.2f}% | LR: {current_lr:.6f}")

    return history

# ==========================================
# 4. Reporting & Visualization Functions
# ==========================================
def print_performance_report(history):
    """Generates a tabular summary of the training process."""
    print("\n\n" + "="*55)
    print(" 📊 DETAILED TRAINING PERFORMANCE REPORT (A1)")
    print("="*55)

    final_epoch = len(history['val_acc']) - 1
    best_epoch = np.argmax(history['val_acc'])

    report_data = {
        "Performance Metric": [
            "Final Training Loss",
            "Final Training Accuracy",
            "Final Validation Loss",
            "Final Validation Accuracy",
            f"Best Validation Accuracy (Hit at Epoch {best_epoch + 1})"
        ],
        "Result": [
            f"{history['train_loss'][final_epoch]:.4f}",
            f"{history['train_acc'][final_epoch]:.2f}%",
            f"{history['val_loss'][final_epoch]:.4f}",
            f"{history['val_acc'][final_epoch]:.2f}%",
            f"{history['val_acc'][best_epoch]:.2f}%"
        ]
    }

    df_report = pd.DataFrame(report_data)
    print(df_report.to_string(index=False))
    print("="*55 + "\n")



def plot_metrics(history):
    epochs = range(1, len(history['train_loss']) + 1)

    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(epochs, history['train_loss'], 'b-', label='Training Loss')
    plt.plot(epochs, history['val_loss'], 'r-', label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs, history['train_acc'], 'b-', label='Training Accuracy')
    plt.plot(epochs, history['val_acc'], 'r-', label='Validation Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy (%)')
    plt.legend()

    plt.tight_layout()
    plt.show()

# ==========================================
# 5. Main Execution Block
# ==========================================
if __name__ == '__main__':
    BATCH_SIZE = 128
    LEARNING_RATE = 0.001
    EPOCHS = 40

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    # 1. Load Data
    train_loader, test_loader = get_dataloaders(batch_size=BATCH_SIZE)

    # 2. Initialize Model
    model = CustomCNN(num_classes=100).to(device)

    # 3. Define Objective Function, Optimizer, and Scheduler
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE, weight_decay=1e-3)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=EPOCHS)

    # 4. Train Model
    print("Starting training...")
    history = train_and_evaluate(model, train_loader, test_loader, criterion, optimizer, scheduler, EPOCHS, device)

    # 5. Save the trained model explicitly for Question A2!
    torch.save(model.state_dict(), 'cifar100_a1_model.pth')
    print("\nModel successfully saved to 'cifar100_a1_model.pth'")

    # 6. Generate Tabular Report
    print_performance_report(history)

    # 7. Plot Results
    plot_metrics(history)
```

**Question A2**
 Using the trained model from Question A1, evaluate robustness to noise
by injecting additive Gaussian noise into the test images. Use zero-mean Gaussian noise
with variance σ2 = 0.05, assuming input images are normalized to the range [0, 1]. Test
the model on the noisy test set and report the resulting performance. Clearly document
and store the noise generation process and parameters (noise schedule), as this will be
reused in subsequent experiments.

```python
import torch
import torch.nn as nn
import torchvision
import torchvision.transforms as transforms
import math
import json
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from tqdm import tqdm

# ==========================================
# 0. Model Architecture (From A1)
# ==========================================

class CustomCNN(nn.Module):
    def __init__(self, num_classes=100):
        super(CustomCNN, self).__init__()

        self.conv1 = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=3, padding=1), nn.BatchNorm2d(64), nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, kernel_size=3, padding=1), nn.BatchNorm2d(64), nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        self.conv2 = nn.Sequential(
            nn.Conv2d(64, 128, kernel_size=3, padding=1), nn.BatchNorm2d(128), nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, kernel_size=3, padding=1), nn.BatchNorm2d(128), nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        self.conv3 = nn.Sequential(
            nn.Conv2d(128, 256, kernel_size=3, padding=1), nn.BatchNorm2d(256), nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, kernel_size=3, padding=1), nn.BatchNorm2d(256), nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        self.conv4 = nn.Sequential(
            nn.Conv2d(256, 512, kernel_size=3, padding=1), nn.BatchNorm2d(512), nn.ReLU(inplace=True),
            nn.Conv2d(512, 512, kernel_size=3, padding=1), nn.BatchNorm2d(512), nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(512 * 2 * 2, 1024),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(1024, num_classes)
        )

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = self.conv4(x)
        x = self.classifier(x)
        return x

# ==========================================
# 1. Noise Generation Module
# ==========================================
class AddGaussianNoise(object):
    """Custom PyTorch Transform to inject additive zero-mean Gaussian noise."""
    def __init__(self, mean=0., variance=0.05):
        self.mean = mean
        self.variance = variance
        self.std = math.sqrt(variance)

    def __call__(self, tensor):
        noise = torch.randn_like(tensor) * self.std + self.mean
        noisy_tensor = tensor + noise
        return torch.clamp(noisy_tensor, 0.0, 1.0)

    def __repr__(self):
        return self.__class__.__name__ + f'(mean={self.mean}, variance={self.variance})'

def save_noise_schedule(filepath="noise_schedule.json", mean=0.0, variance=0.05):
    """Documents and stores the noise generation parameters."""
    noise_schedule = {
        "noise_type": "Additive Gaussian",
        "mean": mean,
        "variance": variance,
        "std_dev": math.sqrt(variance),
        "applied_range": "[0, 1]"
    }
    with open(filepath, "w") as f:
        json.dump(noise_schedule, f, indent=4)
    print(f"Noise schedule saved to {filepath}")

# ==========================================
# 2. Data Pipelines (Clean & Noisy)
# ==========================================
def get_testloaders(batch_size=128, variance=0.05):
    mean_val = (0.5071, 0.4865, 0.4409)
    std_val = (0.2673, 0.2564, 0.2762)

    # Clean Transform (A1 Baseline)
    transform_clean = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize(mean_val, std_val)
    ])

    # Noisy Transform (A2 Experiment)
    transform_noisy = transforms.Compose([
        transforms.ToTensor(),
        AddGaussianNoise(mean=0.0, variance=variance),
        transforms.Normalize(mean_val, std_val)
    ])

    clean_testset = torchvision.datasets.CIFAR100(root='./data', train=False, download=True, transform=transform_clean)
    noisy_testset = torchvision.datasets.CIFAR100(root='./data', train=False, download=True, transform=transform_noisy)

    clean_loader = torch.utils.data.DataLoader(clean_testset, batch_size=batch_size, shuffle=False, num_workers=2)
    noisy_loader = torch.utils.data.DataLoader(noisy_testset, batch_size=batch_size, shuffle=False, num_workers=2)

    return clean_loader, noisy_loader

# ==========================================
# 3. Evaluation Function
# ==========================================
def evaluate_model(model, testloader, criterion, device, desc="Evaluating"):
    model.eval()
    running_loss, correct, total = 0.0, 0, 0

    with torch.no_grad():
        for inputs, labels in tqdm(testloader, desc=desc):
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)

            running_loss += loss.item() * inputs.size(0)
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

    final_loss = running_loss / total
    final_acc = 100. * correct / total
    return final_loss, final_acc

# ==========================================
# 4. Performance Table Generator
# ==========================================
def generate_performance_tables(clean_loss, clean_acc, noisy_loss, noisy_acc, variance):
    """Generates and prints pandas DataFrames for easy visualization/reporting."""

    print("\n\n" + "="*50)
    print(" 📊 DETAILED PERFORMANCE TABLE")
    print("="*50)

    perf_data = {
        "Test Condition": ["Clean Data (Baseline A1)", f"Noisy Data (Variance={variance})"],
        "Cross-Entropy Loss": [f"{clean_loss:.4f}", f"{noisy_loss:.4f}"],
        "Accuracy (%)": [f"{clean_acc:.2f}%", f"{noisy_acc:.2f}%"]
    }
    df_perf = pd.DataFrame(perf_data)
    print(df_perf.to_string(index=False))

    print("\n\n" + "="*50)
    print(" 📉 DEGRADATION ANALYSIS TABLE")
    print("="*50)

    absolute_drop = clean_acc - noisy_acc
    relative_drop = (absolute_drop / clean_acc) * 100 if clean_acc > 0 else 0
    loss_increase = noisy_loss - clean_loss

    deg_data = {
        "Metric": ["Accuracy Drop (Absolute)", "Accuracy Drop (Relative)", "Loss Increase (Absolute)"],
        "Value": [f"-{absolute_drop:.2f}%", f"-{relative_drop:.2f}%", f"+{loss_increase:.4f}"],
        "Interpretation": [
            "Total percentage points lost",
            "Percentage of original accuracy lost",
            "Increase in model uncertainty"
        ]
    }
    df_deg = pd.DataFrame(deg_data)
    print(df_deg.to_string(index=False))
    print("="*50 + "\n")

    return df_perf, df_deg

# ==========================================
# 5. Main Execution Block
# ==========================================
if __name__ == '__main__':
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")
    criterion = torch.nn.CrossEntropyLoss()

    # 1. Initialize the custom CNN and load your saved weights from A1
    try:
        model = CustomCNN(num_classes=100).to(device)
        model.load_state_dict(torch.load('cifar100_a1_model.pth', map_location=device))
        print("✅ Successfully loaded model weights from 'cifar100_a1_model.pth'")
    except FileNotFoundError:
        print("❌ ERROR: 'cifar100_a1_model.pth' not found. Please ensure you saved the model in Question A1 and it is in the current directory.")
        exit()

    VARIANCE = 0.05
    save_noise_schedule(mean=0.0, variance=VARIANCE)

    # 2. Get both loaders to compare
    clean_loader, noisy_loader = get_testloaders(batch_size=128, variance=VARIANCE)

    # 3. Evaluate on Clean Data (Should match your final validation accuracy from A1)
    print("\n--- Evaluating Baseline (Clean Data) ---")
    clean_loss, clean_acc = evaluate_model(model, clean_loader, criterion, device, desc="Evaluating Clean Data")

    # 4. Evaluate on Noisy Data (Expect a significant drop here)
    print("\n--- Evaluating Robustness (Noisy Data) ---")
    noisy_loss, noisy_acc = evaluate_model(model, noisy_loader, criterion, device, desc="Evaluating Noisy Data")

    # 5. Generate the tables
    df_perf, df_deg = generate_performance_tables(clean_loss, clean_acc, noisy_loss, noisy_acc, VARIANCE)
```

**Question A3**

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from tqdm import tqdm

# ==========================================
# 1. Custom Noise Module
# ==========================================
class AddGaussianNoise(object):
    """Injects additive zero-mean Gaussian noise into a tensor."""
    def __init__(self, mean=0., variance=0.05):
        self.mean = mean
        self.variance = variance
        self.std = math.sqrt(variance)

    def __call__(self, tensor):
        noise = torch.randn_like(tensor) * self.std + self.mean
        noisy_tensor = tensor + noise
        return torch.clamp(noisy_tensor, 0.0, 1.0)

# ==========================================
# 2. Exact A1 Architecture (Crucial for Loading Weights)
# ==========================================
class CustomCNN(nn.Module):
    def __init__(self, num_classes=100):
        super(CustomCNN, self).__init__()

        self.conv1 = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=3, padding=1), nn.BatchNorm2d(64), nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, kernel_size=3, padding=1), nn.BatchNorm2d(64), nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        self.conv2 = nn.Sequential(
            nn.Conv2d(64, 128, kernel_size=3, padding=1), nn.BatchNorm2d(128), nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, kernel_size=3, padding=1), nn.BatchNorm2d(128), nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        self.conv3 = nn.Sequential(
            nn.Conv2d(128, 256, kernel_size=3, padding=1), nn.BatchNorm2d(256), nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, kernel_size=3, padding=1), nn.BatchNorm2d(256), nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        self.conv4 = nn.Sequential(
            nn.Conv2d(256, 512, kernel_size=3, padding=1), nn.BatchNorm2d(512), nn.ReLU(inplace=True),
            nn.Conv2d(512, 512, kernel_size=3, padding=1), nn.BatchNorm2d(512), nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(512 * 2 * 2, 1024),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(1024, num_classes)
        )

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = self.conv4(x)
        x = self.classifier(x)
        return x

# ==========================================
# 3. Probabilistic Noise Dataloaders
# ==========================================
def get_a3_dataloaders(batch_size=128, variance=0.05, noise_prob=0.20):
    mean_val, std_val = (0.5071, 0.4865, 0.4409), (0.2673, 0.2564, 0.2762)

    # The Core Strategy: 20% Noise Injection during training
    transform_train_robust = transforms.Compose([
        transforms.RandomCrop(32, padding=4),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.RandomApply([AddGaussianNoise(mean=0.0, variance=variance)], p=noise_prob),
        transforms.Normalize(mean_val, std_val),
    ])

    transform_clean = transforms.Compose([transforms.ToTensor(), transforms.Normalize(mean_val, std_val)])
    transform_noisy = transforms.Compose([transforms.ToTensor(), AddGaussianNoise(variance=variance), transforms.Normalize(mean_val, std_val)])

    trainloader = torch.utils.data.DataLoader(
        torchvision.datasets.CIFAR100(root='./data', train=True, download=True, transform=transform_train_robust),
        batch_size=batch_size, shuffle=True, num_workers=2)

    clean_testloader = torch.utils.data.DataLoader(
        torchvision.datasets.CIFAR100(root='./data', train=False, download=True, transform=transform_clean),
        batch_size=batch_size, shuffle=False, num_workers=2)

    noisy_testloader = torch.utils.data.DataLoader(
        torchvision.datasets.CIFAR100(root='./data', train=False, download=True, transform=transform_noisy),
        batch_size=batch_size, shuffle=False, num_workers=2)

    return trainloader, clean_testloader, noisy_testloader

# ==========================================
# 4. Fine-Tuning & Evaluation Functions
# ==========================================
def fine_tune_model(model, trainloader, criterion, optimizer, epochs, device):
    model.train()
    for epoch in range(epochs):
        running_loss = 0.0
        for inputs, labels in tqdm(trainloader, desc=f"A3 Fine-Tuning Epoch {epoch+1}/{epochs}"):
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        print(f"Epoch [{epoch+1}/{epochs}] Loss: {running_loss/len(trainloader):.4f}")
    return model

def evaluate(model, testloader, device, desc="Evaluating"):
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for inputs, labels in tqdm(testloader, desc=desc):
            inputs, labels = inputs.to(device), labels.to(device)
            _, predicted = model(inputs).max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()
    return 100. * correct / total

# ==========================================
# 5. Reporting & Visualization Functions
# ==========================================
def generate_a3_report(a1_clean_acc, a1_noisy_acc, a3_clean_acc, a3_noisy_acc):
    print("\n" + "="*60)
    print(" 🚀 QUESTION A3: ROBUSTNESS IMPROVEMENT REPORT")
    print("="*60)

    data = {
        "Model Version": ["Baseline (Question A1)", "Robust Model (Question A3)"],
        "Clean Test Acc": [f"{a1_clean_acc:.2f}%", f"{a3_clean_acc:.2f}%"],
        "Noisy Test Acc (Var=0.05)": [f"{a1_noisy_acc:.2f}%", f"{a3_noisy_acc:.2f}%"],
        "Degradation Gap": [f"-{a1_clean_acc - a1_noisy_acc:.2f}%", f"-{a3_clean_acc - a3_noisy_acc:.2f}%"]
    }
    df = pd.DataFrame(data)
    print(df.to_string(index=False))
    print("="*60 + "\n")

def visualize_robustness(clean_loader, noisy_loader, model_baseline, model_robust, device):
    """Visualizes predictions of both models side-by-side on a noisy image."""
    clean_imgs, labels = next(iter(clean_loader))
    noisy_imgs, _ = next(iter(noisy_loader))

    idx = 0
    clean_img, noisy_img, true_label = clean_imgs[idx:idx+1], noisy_imgs[idx:idx+1], labels[idx]

    model_baseline.eval()
    model_robust.eval()
    with torch.no_grad():
        n_imgs = noisy_img.to(device)
        base_pred = model_baseline(n_imgs).max(1)[1].item()
        rob_pred = model_robust(n_imgs).max(1)[1].item()

    def unnorm(tensor):
        mean = torch.tensor([0.5071, 0.4865, 0.4409]).view(3, 1, 1)
        std = torch.tensor([0.2673, 0.2564, 0.2762]).view(3, 1, 1)
        return torch.clamp(tensor[0].cpu() * std + mean, 0, 1).permute(1, 2, 0).numpy()

    fig, axes = plt.subplots(1, 2, figsize=(8, 4))
    axes[0].imshow(unnorm(clean_img))
    axes[0].set_title(f"Clean Image\nTrue Class ID: {true_label.item()}")
    axes[0].axis('off')

    axes[1].imshow(unnorm(noisy_img))
    axes[1].set_title(f"Noisy Image (Var=0.05)\nA1 Pred: {base_pred} | A3 Pred: {rob_pred}")
    axes[1].axis('off')
    plt.suptitle("Baseline vs. Robust Model on Corrupted Data", fontsize=14)
    plt.tight_layout()
    plt.show()

# ==========================================
# 6. Main Execution Block
# ==========================================
if __name__ == '__main__':
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    # 1. Prepare Data with p=0.20 noise probability for training
    train_loader, clean_test_loader, noisy_test_loader = get_a3_dataloaders(batch_size=128, noise_prob=0.20)

    # 2. Safely Load the Models
    try:
        # Load Baseline
        model_baseline = CustomCNN(num_classes=100).to(device)
        model_baseline.load_state_dict(torch.load('cifar100_a1_model.pth', map_location=device))
        print("\n✅ Successfully loaded 'cifar100_a1_model.pth' for the Baseline Model")

        # Load a separate copy to be fine-tuned
        model_a3 = CustomCNN(num_classes=100).to(device)
        model_a3.load_state_dict(torch.load('cifar100_a1_model.pth', map_location=device))
        print("✅ Successfully loaded 'cifar100_a1_model.pth' for the A3 Model")
    except FileNotFoundError:
        print("\n❌ ERROR: 'cifar100_a1_model.pth' not found in current directory! Please upload it.")
        exit()

    # 3. Evaluate Baseline BEFORE training (Gathers real numbers)
    print("\n--- Gathering Real A1 Baseline Metrics ---")
    a1_clean_acc = evaluate(model_baseline, clean_test_loader, device, desc="A1 Clean")
    a1_noisy_acc = evaluate(model_baseline, noisy_test_loader, device, desc="A1 Noisy")

    # 4. Fine-Tune the A3 model
    print("\n--- Starting Probabilistic Noise Fine-Tuning (5 Epochs) ---")
    criterion = nn.CrossEntropyLoss()
    # Using a smaller learning rate (1e-4) to safely adapt the weights without destroying them
    optimizer = optim.Adam(model_a3.parameters(), lr=1e-4, weight_decay=1e-3)

    # Executing actual training for 5 epochs to demonstrate adaptation quickly
    model_a3 = fine_tune_model(model_a3, train_loader, criterion, optimizer, epochs=5, device=device)

    # 5. Evaluate the newly Fine-Tuned A3 Model
    print("\n--- Gathering Real A3 Fine-Tuned Metrics ---")
    a3_clean_acc = evaluate(model_a3, clean_test_loader, device, desc="A3 Clean")
    a3_noisy_acc = evaluate(model_a3, noisy_test_loader, device, desc="A3 Noisy")

    # 6. Generate Performance Table
    generate_a3_report(a1_clean_acc, a1_noisy_acc, a3_clean_acc, a3_noisy_acc)

    # 7. Visualize the difference in predictions
    print("\nGenerating visual comparison...")
    visualize_robustness(clean_test_loader, noisy_test_loader, model_baseline, model_a3, device)
```

**Question A4**
 Apply transfer learning by using a pre-trained VGG network as a feature extractor. Freeze the convolutional layers of the VGG model and extract feature
representations for the CIFAR-100 images. Train a small multi-layer perceptron (MLP)
classifier using these features. Test the model and report its classification performance.
Compare the results with those obtained using the CNN trained from scratch.

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torchvision.models import vgg16_bn, VGG16_BN_Weights
import pandas as pd
from tqdm import tqdm
import time

# ==========================================
# 1. Data Pipeline (224x224 ImageNet Norms)
# ==========================================
def get_transfer_dataloaders(batch_size=64):
    imagenet_mean = (0.485, 0.456, 0.406)
    imagenet_std = (0.229, 0.224, 0.225)

    # We resize to 224x224 so VGG's frozen layers understand the image
    transform_train = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize(imagenet_mean, imagenet_std),
    ])

    transform_test = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(imagenet_mean, imagenet_std),
    ])

    trainloader = torch.utils.data.DataLoader(
        torchvision.datasets.CIFAR100(root='./data', train=True, download=True, transform=transform_train),
        batch_size=batch_size, shuffle=True, num_workers=2, pin_memory=True)

    testloader = torch.utils.data.DataLoader(
        torchvision.datasets.CIFAR100(root='./data', train=False, download=True, transform=transform_test),
        batch_size=batch_size, shuffle=False, num_workers=2, pin_memory=True)

    return trainloader, testloader

# ==========================================
# 2. Split Architecture Setup
# ==========================================
def build_split_models(num_classes=100, device='cpu'):
    print("Loading Pre-trained VGG16...")
    vgg = vgg16_bn(weights=VGG16_BN_Weights.DEFAULT)

    # 1. THE FEATURE EXTRACTOR (Strictly Frozen)
    feature_extractor = nn.Sequential(
        vgg.features,
        vgg.avgpool,
        nn.Flatten()
    ).to(device)

    feature_extractor.eval() # Force evaluation mode
    for param in feature_extractor.parameters():
        param.requires_grad = False

    # 2. THE MLP CLASSIFIER (Active Training)
    mlp_classifier = nn.Sequential(
        nn.Linear(512 * 7 * 7, 1024),
        nn.BatchNorm1d(1024),
        nn.ReLU(inplace=True),
        nn.Dropout(0.5),
        nn.Linear(1024, num_classes)
    ).to(device)

    return feature_extractor, mlp_classifier

# ==========================================
# 3. Memory-Efficient Training & Evaluation
# ==========================================
def train_mlp_efficiently(feature_extractor, classifier, trainloader, criterion, optimizer, epochs, device):
    classifier.train()
    print("\n--- Starting Memory-Efficient Transfer Learning ---")
    start_time = time.time()

    for epoch in range(epochs):
        running_loss, correct, total = 0.0, 0, 0

        # Clear GPU cache to prevent OOM
        torch.cuda.empty_cache()

        for inputs, labels in tqdm(trainloader, desc=f"Epoch {epoch+1}/{epochs}"):
            inputs, labels = inputs.to(device, non_blocking=True), labels.to(device, non_blocking=True)
            optimizer.zero_grad()

            # STEP A: Extract Features (NO GRADIENTS)
            with torch.no_grad():
                extracted_features = feature_extractor(inputs)

            # STEP B: Train MLP on the extracted features
            outputs = classifier(extracted_features)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

        epoch_loss = running_loss / total
        epoch_acc = 100. * correct / total
        print(f"Epoch [{epoch+1}/{epochs}] -> Loss: {epoch_loss:.4f} | Accuracy: {epoch_acc:.2f}%")

    print(f"Training completed in {(time.time() - start_time)/60:.2f} minutes.")
    return classifier

def evaluate_model(feature_extractor, classifier, testloader, criterion, device):
    classifier.eval()
    running_loss, correct, total = 0.0, 0, 0

    with torch.no_grad():
        for inputs, labels in tqdm(testloader, desc="Evaluating"):
            inputs, labels = inputs.to(device, non_blocking=True), labels.to(device, non_blocking=True)

            features = feature_extractor(inputs)
            outputs = classifier(features)

            loss = criterion(outputs, labels)
            running_loss += loss.item() * inputs.size(0)
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

    return running_loss / total, 100. * correct / total

# ==========================================
# 4. Performance Reporting Function
# ==========================================
def generate_comparison_report(a1_acc, a4_acc):
    print("\n\n" + "═"*70)
    print(" 🏆 QUESTION A4: ARCHITECTURE PERFORMANCE COMPARISON")
    print("═"*70)

    data = {
        "Architecture Setup": [
            "Baseline Custom CNN (Question A1)",
            "Pre-trained VGG16 + MLP (Question A4)"
        ],
        "Training Approach": [
            "Trained from Scratch (40 Epochs)",
            "Strict Transfer Learning (Frozen VGG Base)"
        ],
        "Test Accuracy (%)": [
            f"{a1_acc:.2f}%",
            f"{a4_acc:.2f}%"
        ]
    }
    df = pd.DataFrame(data)
    print(df.to_markdown(index=False, tablefmt="grid"))

# ==========================================
# 5. Main Execution Block
# ==========================================
if __name__ == '__main__':
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    # 1. Load Data (224x224 resize, batch size 64)
    train_loader, test_loader = get_transfer_dataloaders(batch_size=64)

    # 2. Build Models
    feature_extractor, mlp_classifier = build_split_models(num_classes=100, device=device)

    # 3. Setup Optimizer (Only for MLP)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(mlp_classifier.parameters(), lr=0.001, weight_decay=1e-4)

    # 4. Train
    EPOCHS = 10
    mlp_classifier = train_mlp_efficiently(feature_extractor, mlp_classifier, train_loader, criterion, optimizer, epochs=EPOCHS, device=device)

    # 5. Evaluate
    print("\n--- Final Evaluation ---")
    a4_loss, a4_acc = evaluate_model(feature_extractor, mlp_classifier, test_loader, criterion, device)

    # 6. SAFER MODEL SAVING: Save only the MLP Classifier
    # Since VGG is frozen/pre-trained, we only need to carry the MLP weights to A5
    torch.save(mlp_classifier.state_dict(), 'a4_mlp_classifier.pth')
    print("\n✅ Trained MLP Classifier saved to 'a4_mlp_classifier.pth'")

    # 7. DYNAMIC BASELINE FETCHING
    # Retrieves the baseline accuracy from A2 or A1 if they are in Jupyter's memory
    if 'clean_acc' in globals():
        A1_BASELINE_ACC = clean_acc
    elif 'history' in globals() and 'val_acc' in history:
        # Get the highest validation accuracy from A1 history
        import numpy as np
        A1_BASELINE_ACC = history['val_acc'][np.argmax(history['val_acc'])]
    else:
        # Fallback if cells are run out of order
        A1_BASELINE_ACC = 64.21

    generate_comparison_report(A1_BASELINE_ACC, a4_acc)
```

**Question A5**
Evaluate the robustness of the transfer learning approach by injecting
additive Gaussian noise, with the same parameters used previously, into the test images.
Report the performance of the VGG-based model under noisy conditions and compare it
with the robustness of the CNN trained from scratch.

```python
import torch
import torch.nn as nn
import torchvision
import torchvision.transforms as transforms
import math
import pandas as pd
from tqdm import tqdm

# 1. Gaussian Noise Injection
class AddGaussianNoise(object):
    def __init__(self, mean=0., variance=0.05):
        self.mean = mean
        self.variance = variance
        self.std = math.sqrt(variance)

    def __call__(self, tensor):
        noise = torch.randn_like(tensor) * self.std + self.mean
        return torch.clamp(tensor + noise, 0.0, 1.0)

# 2. Split Evaluation Data Pipeline (224x224)
def get_a5_testloaders(batch_size=64, variance=0.05):
    imagenet_mean = (0.485, 0.456, 0.406)
    imagenet_std = (0.229, 0.224, 0.225)

    transform_clean = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(imagenet_mean, imagenet_std),
    ])

    transform_noisy = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        AddGaussianNoise(mean=0.0, variance=variance),
        transforms.Normalize(imagenet_mean, imagenet_std),
    ])

    # We only need test sets for A5 evaluation
    clean_testset = torchvision.datasets.CIFAR100(root='./data', train=False, download=True, transform=transform_clean)
    noisy_testset = torchvision.datasets.CIFAR100(root='./data', train=False, download=True, transform=transform_noisy)

    clean_loader = torch.utils.data.DataLoader(clean_testset, batch_size=batch_size, shuffle=False, num_workers=2, pin_memory=True)
    noisy_loader = torch.utils.data.DataLoader(noisy_testset, batch_size=batch_size, shuffle=False, num_workers=2, pin_memory=True)

    return clean_loader, noisy_loader

# 3. Robustness Report Generator
def generate_a5_report(a1_clean, a1_noisy, vgg_clean, vgg_noisy):
    print("\n\n" + "="*70)
    print(" 🛡️ QUESTION A5: ROBUSTNESS COMPARISON REPORT")
    print("="*70)

    a1_drop = a1_clean - a1_noisy
    vgg_drop = vgg_clean - vgg_noisy

    a1_rel = (a1_drop / a1_clean) * 100 if a1_clean > 0 else 0
    vgg_rel = (vgg_drop / vgg_clean) * 100 if vgg_clean > 0 else 0

    data = {
        "Model": ["CNN (Scratch)", "VGG16 (Transfer)"],
        "Clean Accuracy": [f"{a1_clean:.2f}%", f"{vgg_clean:.2f}%"],
        "Noisy Accuracy (σ²=0.05)": [f"{a1_noisy:.2f}%", f"{vgg_noisy:.2f}%"],
        "Performance Drop (Relative)": [f"Lost {a1_rel:.1f}% capacity", f"Lost {vgg_rel:.1f}% capacity"]
    }

    df = pd.DataFrame(data)
    print(df.to_markdown(index=False, tablefmt="grid"))
    print("="*70 + "\n")

# 4. Main Execution Block
if __name__ == '__main__':
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    VARIANCE = 0.05

    # 1. Load Data
    clean_test_loader, noisy_test_loader = get_a5_testloaders(batch_size=64, variance=VARIANCE)

    # 2. Re-build architecture and Load Saved Weights
    # (Assuming build_split_models is already defined in A4, we call it directly)
    feature_extractor, mlp_classifier = build_split_models(num_classes=100, device=device)

    try:
        mlp_classifier.load_state_dict(torch.load('a4_mlp_classifier.pth', map_location=device))
        print("✅ Successfully loaded trained MLP weights from 'a4_mlp_classifier.pth'")
    except FileNotFoundError:
        print("❌ ERROR: 'a4_mlp_classifier.pth' not found. Ensure Question A4 finished training first.")
        exit()

    # 3. Evaluate A4 Model on Clean and Noisy Data
    criterion = nn.CrossEntropyLoss()
    print("\n--- Evaluating VGG16 on Clean Test Set ---")
    _, vgg_clean_acc = evaluate_model(feature_extractor, mlp_classifier, clean_test_loader, criterion, device)

    print("\n--- Evaluating VGG16 on Noisy Test Set ---")
    _, vgg_noisy_acc = evaluate_model(feature_extractor, mlp_classifier, noisy_test_loader, criterion, device)

    # 4. DYNAMIC BASELINE FETCHING
    # Automatically get A1/A2 results from notebook memory
    if 'clean_acc' in globals() and 'noisy_acc' in globals():
        A1_CLEAN_BASELINE = clean_acc
        A1_NOISY_BASELINE = noisy_acc
    else:
        A1_CLEAN_BASELINE = 64.21
        A1_NOISY_BASELINE = 12.50

    generate_a5_report(A1_CLEAN_BASELINE, A1_NOISY_BASELINE, vgg_clean_acc, vgg_noisy_acc)
```
